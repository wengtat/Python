# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GGL','AMZN','MSFT','FB']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
plt.bar(company,revenue)
```

```python
# yposition
xpos = np.arange(len(company))
xpos
```

```python
# Xticks replace numerical array value with the corresponding x axis list
plt.xticks(ypos,company)
plt.ylabel('Revenue')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2,revenue, width=0.4, label ='Revenue')
plt.bar(xpos+0.2,profit,width=0.4, label = 'Profit')
plt.legend()
```

```python
# Horizontal bar chart
plt.yticks(ypos,company)
plt.title('US Tech Stocks')

plt.barh(xpos-0.2,revenue, label ='Revenue')
plt.barh(xpos+0.2,profit, label = 'Profit')
plt.legend()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar)
#Default pass 10 bins
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color = 'g', histtype = 'step')
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.xlabel('Blood sugar level range')
plt.ylabel('Number of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150],rwidth=0.95,label=['men','women'])
plt.legend()
```

```python
plt.xlabel('Blood sugar level range')
plt.ylabel('Number of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150],rwidth=0.95,label=['men','women'],orientation ='horizontal')
plt.legend()
```

# Pie Chart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities"]
```

```python
plt.pie(exp_vals,labels = exp_labels,autopct="%0.2f%%")
plt.axis("equal")
plt.show()
```

```python
plt.pie(exp_vals,labels = exp_labels,autopct="%0.1f%%", explode=[0,0.1,0,0,0])
plt.show()
```
